```python
import nltk
from nltk import word_tokenize
import string
import re
from nltk import tokenize
from nltk.corpus import stopwords
from collections import Counter
from wordcloud import WordCloud
from nltk import ngrams
```

```python
nltk.download('stopwords')
```

```python
cnt = Counter()
stop = stopwords.words('english')
```

```python
df1['description'] = df1['description'].str.lower()
df1['description'] = df1['description'].str.replace("[^\w\s]"," ")
df1['description'] = df1['description'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop]))
```

```python
df1['description']
```

```python
text = []
for i in df1['description']:
  text.append(i.split())
words = []
for i in text:
  for j in i:
    words.append(j)
fdist = nltk.FreqDist(words)
print(fdist.most_common(10))
```

```python
plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
keys = [i[0] for i in fdist.most_common(10)]
values = [i[1] for i in fdist.most_common(10)]
sns.barplot(keys, values)

plt.subplot(1, 2, 2)
plt.pie(values, labels=keys, autopct="%0.2f%%")
```
